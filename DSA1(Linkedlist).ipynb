### INSERTION AT ENDING

```python
class Node:
    def __init__(self , new_data):
        self.data = new_data
        self.next = None 

class Linkedlist:
    def __init__(self):
        self.head = None 

    def InsertAtEnd(self , new_data):
        new_node = Node(new_data)

        if self.head is None:
            self.head = new_node
            return

        current = self.head
        while current.next is not None:
            current = current.next
            
        current.next = new_node

    # INSERION AT GIVEN INDEX 
    
    def InsertAtGivenIndex(self , index , new_data):
        new_node = Node(new_data)

        if index == 0:
            new_node.next = self.head
            self.head = new_node
            return

        current = self.head
        current_index = 0

        while current is not None and current_index < index - 1:
            current = current.next
            current_index += 1

        if current is None:
            print('Index out of range')
            return

        new_node.next = current.next
        current.next = new_node

    def display(self):
        current = self.head
        while current:
            print(current.data , end = ' --> ')
            current = current.next
        print('None')

li = Linkedlist()
li.InsertAtEnd(11)
li.InsertAtEnd(12)
li.InsertAtEnd(13)
li.InsertAtEnd(14)

li.display()

li.InsertAtGivenIndex(3 , 100)
li.display()

li.InsertAtGivenIndex(8 , 200)
li.display()
        
```

### REMOVE NODE AT SPECIFIC INDEX

```python
class Node:
    def __init__(self , new_data):
        self.data = new_data
        self.next = None

class Linkedlist:
    def __init__(self):
        self.head = None

    def remove_index(self , index):
        #CASE - 1 : LIST IS EMPTY
        if self.head is None:
            print('List is empty')
            return 

        #CASE - 2 : REMOVE HEAD (INDEX = 0)
        if index == 0:
            self.head = self.head.next

        #CASE - 3 : TRAVERSE UNTIL (INDEX - 1)
        current = self.head
        current_index = 0
        
        while current is not None and current_index < index - 1:
            current = current.next
            current_index += 1

        #CASE - 4 : INDEX IS OUT OF RANGE
        if current.next is None:
            print('The index is out of range')
            return

        current.next = current.next.next
        print(f'Node at {index} is removed')

    def display(self):
        current = self.head
        while current is not None:
            print(current.data , end = ' --> ')
            current = current.next
        print('None')

li = Linkedlist()
li.head = Node(10)
li.head.next = Node(20)
li.head.next.next = Node(30)
li.head.next.next.next = Node(40)

print('Before deleting index : ')
li.display()

li.remove_index(2)
print('After removing the index : ')

li.display()
        
```

```python
class Node:
    def __init__(self , new_data):
        self.data = new_data
        self.next = None

class Linkedlist:
    def __init__(self):
        self.head = None

    def remove_specific(self , value):

        #CASE - 1 : LIST IS EMPTY
        if self.head is None:
            print('The list is empty')
            return

        #CASE - 2 : IF THE HEAD ITSELF IS THE TARGET
        if self.head.data == value:
            self.head = self.head.next
            print(f'Node {value} is removed')
            return

        #CASE - 3 : TRAVERSE TO FIND THE NODE BEFORE TARGET
        current = self.head
        while current is not None and current.next.data != value:
            current = current.next

        if current.next is None:
            print('The target is not found')
            return

        current.next = current.next.next
        print(f'Node {value} is removed')

    def display(self):
        current = self.head
        while current is not None:
            print(current.data , end = ' --> ')
            current = current.next
        print('None')

li = Linkedlist()
li.head = Node(10)
li.head.next = Node(20)
li.head.next.next = Node(30)
li.head.next.next.next = Node(40)

print('Before deletion : ')
li.display()

li.remove_specific(20)
print('After deletion : ')

li.display()
```

### REVERSE LINKED LIST (ITERATIVE METHOD)

```python
class Node:
    def __init__(self , new_data):
        self.data = new_data
        self.next = None

class Linkedlist:
    def __init__(self):
        self.head = None

    def reverse(self):
        current = self.head
        prev = None

        while current is not None:
            next_node = current.next
            current.next = prev
            prev = current
            current = next_node

        self.head = prev

    def display(self):
        current = self.head
        while current is not None:
            print(current.data , end = ' --> ')
            current = current.next
        print('None')

li = Linkedlist()
li.head = Node(10)
li.head.next = Node(20)
li.head.next.next = Node(30)
li.head.next.next.next = Node(40)

print('List before reversing : ')
li.display()

li.reverse()
print('List after reversing : ')
li.display()
```

### REVERSING LINKEDLIST (RECURSION)

```python
class Node:
    def __init__(self , new_data):
        self.data = new_data
        self.next = None

class Linkedlist:
    def __init__(self):
        self.head = None

    def reverse_recursive(self , head):
        # BASE CASE:
        # if list is empty OR only one node,
        # that node is already reversed, so return it
        if head is None or head.next is None:
            return head

        # RECURSIVE CALL:
        # reverse the rest of the list starting from next node
        # this will return the NEW head (last node of original list)
        new_head = self.reverse_recursive(head.next)


        head.next.next = head

        # BREAK OLD LINK:
        # current node should no longer point forward
        head.next = None

        # RETURN NEW HEAD:
        # always return the same new head (fixed at base case)
        return new_head
    def reverse(self):
        # update the list head using recursive function
        self.head = self.reverse_recursive(self.head)

    def display(self):
        current = self.head
        while current:
            print(current.data, end=" --> ")
            current = current.next
        print("None")

li = Linkedlist()
li.head = Node(10)
li.head.next = Node(20)
li.head.next.next = Node(30)

print("Before reverse:")
li.display()

li.reverse()

print("After reverse:")
li.display()
```

### n-th NODE FROM THE END

```python
class Node:
    def __init__(self , new_data):
        self.data = new_data  # store value
        self.next = None      # pointer to next node

class Linkedlist:
    def __init__(self):
        self.head = None   # start of the list
     
    # Function to display the linked list
    def display(self):
        current = self.head
        while current is not None:
            print(current.data , end = ' --> ')
            current = current.next
        print('None')

    def nth_value(self , n):

        #CASE - 1 : LIST IS EMPTY
        if self.head is None:
            print('The list is empty')
            return

        first = self.head
        second = self.head

        # Step 1: move first pointer n steps ahead
        for _ in range(n):
            if first is None:
                print('n is greater than the length of the list ')
                return

            first = first.next

         # Step 2: move both pointers together
        while first is not None:
            first = first.next
            second = second.next

         # Step 3: second now points to nth node from end
        print(f'{n}th node from the end is : ',second.data)

li = Linkedlist()
li.head = Node(11)
li.head.next = Node(12)
li.head.next.next = Node(13)
li.head.next.next.next = Node(14)
li.head.next.next.next.next = Node(15)

print('The linked list : ')
li.display()

li.nth_value(2)
li.nth_value(4)
```

### DELETE ALL OCCURENCES OF A GIVEN VALUE

```python
class Node:
    def __init__(self , new_data):
        self.data = new_data
        self.next = None

class Linkedlist:
    def __init__(self):
        self.head = None

    def display(self):
        current = self.head
        while current is not None:
            print(current.data , end = ' --> ')
            current = current.next
        print('None')

    def delete_all_givenvalue(self , value):

        current = self.head

         # STEP 1: Remove value from the head (if present)
        while self.head is not None and self.head.data == value:
            self.head = self.head.next
            return

        # STEP 2: Remove value from remaining nodes
        while current is not None and current.next is not None:
            if current.next.data == value:
                current.next = current.next.next   # skip the node
            else:
                current = current.next    # move forward

li = Linkedlist()

li.head = Node(1)
li.head.next = Node(2)
li.head.next.next = Node(7)
li.head.next.next.next = Node(3)
li.head.next.next.next.next = Node(4)
li.head.next.next.next.next.next = Node(7)

print('The linked list : ')
li.display()

print('After deleting 7 : ')
li.delete_all_givenvalue(7)
li.display()
```

### DETECT LOOP IN A LINKED LIST

```python
class Node:
    def __init__(self , new_data):
        self.data = new_data
        self.next = None

class Linkedlist:
    def __init__(self):
        self.head = None

    def display(self):
        current = self.head
        while current is not None:
            print(current.data , end = ' --> ')
            current = current.next
        print('None')

    def detect_loop(self):
        slow = self.head
        fast = self.head

        while fast is not None and fast.next is not None:
            slow = slow.next
            fast = fast.next.next

            if slow == fast:
                print('Loop detected')
                return

        print('Loop doesnt exist')

li = Linkedlist()
li.head = Node(11)
li.head.next = Node(12)
li.head.next.next = Node(13)
li.head.next.next.next = Node(14)
li.head.next.next.next.next = Node(15)

li.head.next.next.next.next = li.head.next


li.detect_loop()


# We use slow and fast pointers. If they meet, a loop exists. If fast reaches null, there is no loop.
```

### MERGING TWO SORTED LISTS

```python
class Node:
    def __init__(self , new_data):
        self.data = new_data
        self.next = None

class Linkedlist:
    def __init__(self):
        self.head = None

    def display(self):
        current = self.head
        while current is not None:
            print(current.data , end = ' --> ')
            current = current.next
        print('None')
        
    def merged(self , head1 , head2):
        dummy = Node(0)
        tail = dummy

        while head1 is not None and head2 is not None:
            if head1.data <= head2.data:
                tail.next = head1
                head1 = head1.next
            else:
                tail.next = head2
                head2 = head2.next

            tail = tail.next

        if head1 is not None:
            tail.next = head1
        else:
            tail.next = head2

        return dummy.next

# First sorted list
l1 = Linkedlist()
l1.head = Node(10)
l1.head.next = Node(20)
l1.head.next.next = Node(40)

# Second sorted list
l2 = Linkedlist()
l2.head = Node(15)
l2.head.next = Node(30)
l2.head.next.next = Node(35)

print('List 1 : ')
l1.display()

print('List 2 : ')
l2.display()

merged_head =l1.merged(l1.head , l2.head)

merge = Linkedlist()

merge.head = merged_head

print('Merged List : ')
merge.display()
```

### REMOVE DUPLICATES FROM A LINKED LIST

```python
class Node:
    def __init__(self , new_data):
        self.data = new_data
        self.next = None

class Linkedlist:
    def __init__(self):
        self.head = None

    def duplicate(self):
        if self.head is None:
            print('The list is empty')
            return

        current = self.head
        seen = set()
        prev = None

        while current is not None:
            if current.data in seen:
                prev.next = current.next
            else:
                seen.add(current.data)
                prev = current

            current = current.next
    def display(self):
        current = self.head
        while current is not None:
            print(current.data , end = ' --> ')
            current = current.next
        print('None')
        
li  = Linkedlist()
values = [1 , 2 , 3 , 5 , 6 , 7 , 7 , 1 , 3 , 5]

li.head = Node(values[0])
current = li.head

for v in values[1:]:
    current.next = Node(v)
    current = current.next
# li.head = Node(1)
# li.head.next = Node(2)
# li.head.next.next = Node(7)
# li.head.next.next.next = Node(3)
# li.head.next.next.next.next = Node(4)
# li.head.next.next.next.next.next = Node(7)

print('Linked list')
li.display()

print('after removing duplicates : ')
li.duplicate()
li.display()
```
