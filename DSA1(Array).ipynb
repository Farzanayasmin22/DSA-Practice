### LEETCODE QUESTION - 53 MAXIMUM SUBARRAY

#### Given an integer array nums, find the subarray with the largest sum, and return its sum.

 

Example 1:

Input: nums = [-2,1,-3,4,-1,2,1,-5,4]
Output: 6
Explanation: The subarray [4,-1,2,1] has the largest sum 6.

```python

nums = [-2,1,-3,4,-1,2,1,-5,4]
current_sum = nums[0]
max_sum = nums[0]

for i in range(1 , len(nums)):

    current_sum = max(nums[i] , current_sum + nums[i])
    max_sum = max(max_sum , current_sum)

print(max_sum)
```

### LEETCODE QSTN - 217 DUPLICATE

#### Given an integer array nums, return true if any value appears at least twice in the array, and return false if every element is distinct.

 

Example 1:

Input: nums = [1,2,3,1]

Output: true


```python
class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:
        dict = {}

        for i in nums:
            if i not in dict:
                dict[i] = 1
            else:
                return True
        return False
```

### LEETCODE QSTN - 1 ADD TWO SUM

```python
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        seen = {}

        for i, num in enumerate(nums):
            diff = target - num
            if diff in seen:
                return [seen[diff], i]
            seen[num] = i
```

### LEETCODE QSTN - 283 MOVING ZEROES AT END

```python
nums = [1 , 0 , 4 , 5 , 0 , 0 , 11]
pos = 0

for i in range(len(nums)):
    if nums[i] != 0:
        nums[pos] = nums[i]
        pos += 1

for i in range(pos , len(nums)):
    nums[i] = 0

print(nums)
```

### BINARY SEARCH USING RECURSION

```python
def binarySearch(nums , left , right , target):
    if left > right:
        return -1

    mid = (right + left) // 2
    if nums[mid] == target:
        return mid
    elif nums[mid] < target:
        return binarySearch(nums , mid + 1 , right , target)
    else:
        return binarySearch(nums , mid - 1 , left , target)

nums = [1 , 3 , 5 , 7 , 9]
target = 7
result = binarySearch(nums , 0 , len(nums) - 1 , target)
print(result)
```

### FINDING FIRST OCCURENCE OF THE TARGET USING BINARY SEARCH

```python
def firstOccurence(nums , target):
    left = 0
    right = len(nums) - 1
    ans = -1

    while left <= right:
        mid = (left + right) // 2
        if nums[mid] == target:
            ans = mid
            right = mid - 1
        elif nums[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return ans
nums = [1 , 1 , 1 , 4 , 5]
target = 1

result = firstOccurence(nums , target)
print(result)                 
            
```

### FINDING LAST OCCURENCE USING RECURSION

```python
def lastOccurence(nums , target):
    left = 0
    right = len(nums) - 1
    ans = -1

    while left <= right:
        mid = (left + right) // 2
        if nums[mid] == target:
            ans = mid
            left = mid + 1
        elif nums[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return ans

nums = [3 , 4 , 4 , 4 , 7 , 9]
target = 4
res = lastOccurence(nums , target)
print(res)
```

### LARGEST AFTER THE TARGET USING BINARY RECURSION

```python
def largest(nums , target):
    left = 0
    right = len(nums) - 1
    ans = -1

    while left <= right:
        mid = (left + right) // 2
        if nums[mid] > target:
            ans = nums[mid]
            right = mid - 1
        else:
            left = mid + 1
    return ans

nums = [1 , 2 , 3 , 4 , 5 , 6]
target = 4
res = largest(nums , target)
print(res)
```

### RECURSION

```python
def fact(n):
    while n == 0:
        return 1
    else:
        return n * fact(n - 1)

print(fact(3))
```

```python
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)

print(fibonacci(4))
```

```python
#sum of n natural numbers

def sum(n):
    if n == 0:
        return 1
    return n + sum(n - 1)

print(sum(5))
```

```python
def reverse(s):
    rev = ''
    for ch in s:
        rev = ch + rev
    return rev
print('Reversed string : ',reverse('FARSANA'))
```

```python
#count vowels

def vowels(s):
    vowels = 'aeiouAEIOU'
    count = 0

    for ch in s:
        if ch in vowels:
            count += 1
    return count

print('vowels : ',vowels('Farsana'))
```

```python
def palindrome(s):
    return s == s[::-1]

print(palindrome('sas'))
print(palindrome('sinu'))
```

```python
def string(s):
    count = 0
    for _ in s:
        count += 1
    return count
print(string('farsana'))
```
