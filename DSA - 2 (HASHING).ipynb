### CHAINING

```python
size = 5
table = [[] for _ in range(size)] 

print(table)
```

### INSERTION

```python
def insert(table , key , value):
    index = hash(key) % len(table)
    table[index].append((key , value))

insert(table , 'apple' , 50)
insert(table , 'mango' , 15)
insert(table , 'grapes' , 40)

print(table)
```

### SEARCHING

```python
 def search(table , key):
     index = hash(key) % len(table)
     for k , v in table[index]:
         if k == key:
             return v
     return None

print(search(table , 'apple'))
```

```python
def delete(table , key):
    index = hash(key) % len(table)

    for i , (k , v) in enumerate(table[index]):
        if k == key:
            del table[index][i]
            return True
    return False

delete(table , 'grapes')
print(table)
```

### HASH TABLE OPERATIONS

```python
size = 5
table = [[] for _ in range(size)]

def insert(table , key , value):
    index = hash(key) % len(table)
    table[index].append((key , value))

def search(table , key):
    index = hash(key) % len(table)
    for k , v in table[index]:
        if k == key:
            return v
    return None

def delete(table , key):
    index = hash(key) % len(table)
    for i , (k , v) in enumerate(table[index]):
        if k == key:
            del table[index][i]
            return True
    return False

insert(table , 'grapes' , 30)
insert(table , 'orange' , 20)
insert(table , 'lemon' , 60)

print('Table after insertion :')
print(table)

print('\nSearch result : ')
print('grapes : ' ,search(table , 'grapes'))
print('apple : ' ,search(table , 'apple'))

delete(table, "grapes")

print('\nTable after deletion : ')
print(table)
```
