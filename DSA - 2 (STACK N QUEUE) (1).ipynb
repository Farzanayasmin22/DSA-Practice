### STACK(BASIC OPERATIONS)

```python
stack = []

def isEmpty():
    return len(stack) == 0

def push(x):
    stack.append(x)

def pop():
    if isEmpty():
        return 'Stack is empty'
    return stack.pop()

def peek():
    if isEmpty():
        return 'Stack is empty'
    return stack[-1] 

def size():
    return len(stack)

push(10)
push(20)
push(30)
push(40)
push(50)

print('Top element : ',peek())
print('Stack size : ',size())
print('Popped : ',pop())
print('Stack after pop : ',stack)
```

### QUEUE (BASIC OPERATIONS)

```python
queue = []

def isEmpty():
    return len(queue) == 0

def enqueue(x):
    queue.append(x)
    print(x , 'added to queue')

def dequeue():
    if isEmpty():
        print('Queue is empty')
    else:
        print(queue.pop(0) , 'is removed')

def front():
    if isEmpty():
        print('Queue is empty')
    else:
        print(queue[0] , 'is first element')

enqueue(10)
enqueue(20)
enqueue(30)
enqueue(40)

front()
dequeue()
front()
print('Queue : ',queue)
```

### DFS (DEPTH FIRST SEARCH)

```python
def dfs(graph , node , visited):
    visited.add(node)
    print(node , end = ' ')

    for neighbour in graph[node]:
        if neighbour not in visited:
            dfs(graph , neighbour , visited)

graph = {
    0: [1, 2],
    1: [0, 3],
    2: [0],
    3: [1]
}

visited = set()
dfs(graph , 0 , visited)
```

### BSF (BREADTH SEARCH FIRST)

```python
from collections import deque

def bfs(graph , start):
    visited = set()
    queue = deque()

    queue.append(start)
    visited.add(start)

    while queue:
        node = queue.popleft()
        print(node , end = ' ')

        for neighbour in graph[node]:
            if neighbour not in visited:
                visited.add(neighbour)
                queue.append(neighbour)

graph = {
    0 : [1 , 2],
    1 : [0 , 3],
    2 : [0],
    3 : [1]
}

bfs(graph , 0)
```

### REVERSE A STRING USING STACK

```python
def reverse(s):
    stack = []
    for ch in s:
        stack.append(ch)

    rev = ''
    while stack:
        rev += stack.pop()

    return rev
    
s = 'Farsana'
print('Reversed string : ',reverse(s))
```

### STACK USING LINKEDLIST

```python
class Node:
    def __init__(self , data):
        self.data = data
        self.next = None

class Stack:
    def __init__(self):
        self.head = None

    #push operation
    def push(self , data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node
        print(data , 'pushed to stack')

    #pop operation
    def pop(self):
        if self.head is None:
            print('The stack is empty')
            return None
        popped = self.head.data
        self.head = self.head.next
        return popped

    def peek(self):
        if self.head is None:
            print('The stack is empty')
            return None
        return self.head.data

    def isEmpty():
        return self.head is None

    def display():
        current = self.head
        while current:
            print(current.data , end = ' --> ')
            current = current.next
        print('None')

s = Stack()
s.push(10)
s.push(20)
s.push(30)
s.push(40)
s.push(50)

display()

print('Popped : ',s.pop())
print('Peek : ',s.peek()) 

display()
```

### IMPLEMENTING STACK USING ONE QUEUE

```python
from collections import deque

class Stack:
    def __init__(self):
        self.q = deque()

    def push(self , x):
        self.q.append(x)

        for _ in range(len(self.q) - 1):
            self.q.append(self.q.popleft())

    def pop(self):  
        if self.is_empty():
            return 'Stack is empty'
        return self.q.popleft()

    def top(self):
        if self.is_empty():
            return 'Stack is empty'
        return self.q[0]

    def is_empty(self):
        return len(self.q) == 0

s = Stack()
s.push(1)
s.push(2)
s.push(3)
s.push(4)

print('Stack : ',s.q)
print(s.pop())
print(s.top())
print(s.pop())
print(s.pop())
print(s.pop())
print(s.pop())
```

### IMPLEMENTING QUEUE USING TWO STACK

```python
class Queue:
    def __init__(self):
        self.s1 = []
        self.s2 = []

    def enqueue(self , x):
        self.s1.append(x)

    def dequeue(self):
        if self.is_empty():
            return 'Queue is empty'

        if not self.s2:
            while self.s1:
                self.s2.append(self.s1.pop())

        return self.s2.pop()

    def is_empty(self):
        return not self.s1 and not self.s2

    def display(self):
        if self.is_empty():
            print("Queue is empty")
        else:
            print("Queue:", self.s2[::-1] + self.s1) #We reverse s2 because its order is backward,
                                                    #then add s1 because it already has correct order.
            
q = Queue()

q.enqueue(1)
q.enqueue(2)
q.enqueue(3)
q.enqueue(4)
q.enqueue(5)

print(q.dequeue())
print(q.dequeue())

q.display()
q.enqueue(6)

print(q.dequeue())
print(q.dequeue())
print(q.dequeue())
print(q.dequeue())
```
